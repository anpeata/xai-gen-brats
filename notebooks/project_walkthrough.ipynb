# XAI-BraTS Project Walkthrough

This notebook demonstrates the full research pipeline for **Explainable & Generative AI for Brain Tumor MRI Analysis** on BraTS.

## 1) Goals
- Train a baseline multi-modal segmentation model
- Evaluate Dice, HD95, and ECE
- Visualize explainability maps (Grad-CAM, SHAP)
- Estimate uncertainty with MC Dropout
- Generate synthetic MRI slices with a VAE

````python
from pathlib import Path
import matplotlib.pyplot as plt

root = Path('..')
assets = root / 'assets'
print('Assets available:', sorted([p.name for p in assets.glob('*.svg')]))
````

````python
# Visual placeholders for presentation before full training runs
fig, ax = plt.subplots(1, 3, figsize=(16, 4))
for i, name in enumerate(['example_prediction.svg', 'gradcam_overlay.svg', 'uncertainty_map.svg']):
    img = plt.imread(assets / name)
    ax[i].imshow(img)
    ax[i].axis('off')
    ax[i].set_title(name.replace('_', ' ').replace('.svg', ''))
plt.tight_layout()
plt.show()
````
